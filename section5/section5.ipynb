```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_dataset.csv")

# Group data
avg_score_gender = df.groupby("Gender")["Final_Exam_Score"].mean()

# Bar chart
avg_score_gender.plot(kind="bar")

plt.title("Average Final Exam Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
# Sort by Hours Studied
df_sorted = df.sort_values(by="Hours_Studied")

# Line chart
plt.plot(df_sorted["Hours_Studied"], df_sorted["Final_Exam_Score"])

plt.title("Hours Studied vs Final Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Final Exam Score")
plt.show()
```
