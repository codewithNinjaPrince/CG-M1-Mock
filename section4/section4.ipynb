```python
import pandas as pd


# 1. Load Dataset

df = pd.read_csv("student_dataset.csv")

print("First 5 rows:\n", df.head())


# 2. Filtering

# Students who studied more than 8 hours
high_study = df[df["Hours_Studied"] > 8]
print("\nStudents who studied > 8 hours:\n", high_study)

# Students who failed
failed_students = df[df["Pass_Fail"] == "Fail"]
print("\nFailed Students:\n", failed_students)

# 3. Grouping

# Average Final Exam Score by Gender
avg_score_gender = df.groupby("Gender")["Final_Exam_Score"].mean()
print("\nAverage Final Exam Score by Gender:\n", avg_score_gender)

# Average GPA by Pass/Fail
avg_gpa_status = df.groupby("Pass_Fail")["Previous_Sem_GPA"].mean()
print("\nAverage GPA by Pass/Fail:\n", avg_gpa_status)

# 4. Sorting
# Top students based on Final Exam Score
top_students = df.sort_values(by="Final_Exam_Score", ascending=False)
print("\nTop Performing Students:\n", top_students.head())


# 5. Insights
# Insight 1: Who performs better (Male/Female)
top_gender = avg_score_gender.idxmax()
print(f"\nInsight 1: {top_gender} students have higher average final exam scores.")

# Insight 2: Top performer
top_student = top_students.iloc[0]
print(f"Insight 2: Top student is {top_student['Student_ID']} with score {top_student['Final_Exam_Score']}")

# Bonus Insight: Study hours impact
study_vs_score = df.groupby("Hours_Studied")["Final_Exam_Score"].mean()
print("\nBonus Insight: Study hours vs score:\n", study_vs_score)
```
